```python
#gender distribution
plt.figure(figsize=(5,5))
a=sns.countplot(data=df,x="Gender")
plt.title("Gender distribution")
a.bar_label(a.containers[0])
plt.show()
```

```python
#number of females in data is more than number of males
```

```python
gb=df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
```

```python
gb.head()
```

```python
sns.heatmap(gb,annot=True)
plt.title("Relationship between parent education and student education")
plt.show()
```

```python
gb1=df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
gb1.head()
```

```python
sns.heatmap(gb1,annot=True)
plt.title("Relationship between parent maritial status and student education")
plt.show()
```

```python
#from the chart we can state that maratial status of parent do not have anny impact on student score.
```

#DIstribution of ethnic group

```python
groupA=df.loc[(df['EthnicGroup']=="group A")].count()
groupB=df.loc[(df['EthnicGroup']=="group B")].count()
groupC=df.loc[(df['EthnicGroup']=="group C")].count()
groupD=df.loc[(df['EthnicGroup']=="group D")].count()
groupE=df.loc[(df['EthnicGroup']=="group E")].count()
mlist=[groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
l=["group A","group B","group C","group D","group E"]
plt.title("Distribution of ethnic group")
plt.pie(mlist,labels=l,autopct="%1.2f%%")
```

```python
ax=sns.countplot(data=df,x="EthnicGroup")
ax.bar_label(ax.containers[0])
```

```python
A=df.loc[(df['WklyStudyHours']=="< 5")].count()
B=df.loc[(df['WklyStudyHours']=="5 - 10")].count()
C=df.loc[(df['WklyStudyHours']== "> 10")].count()

mlist=[A["WklyStudyHours"],B["WklyStudyHours"],C["WklyStudyHours"]]
l=["< 5","5 - 10","> 10"]
plt.title("Distribution of weekly study hour")
plt.pie(mlist,labels=l,autopct="%1.2f%%")
```
